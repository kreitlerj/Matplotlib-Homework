```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
study_data = pd.merge(clinical_trial_data, mouse_drug_data, on='Mouse ID')

# Display the data table for preview
study_data.head()
```

## Tumor Response to Treatment

```python
# Create dataframe to store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_response_mean_df = pd.DataFrame(study_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean())
tumor_response_mean_df = tumor_response_mean_df.reset_index()
# Preview DataFrame
tumor_response_mean_df.head()
```

```python
# Create a dataframe and store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_response_error_df = pd.DataFrame(study_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem())
tumor_response_error_df = tumor_response_error_df.reset_index()
# Preview DataFrame
tumor_response_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_response_mean_df = tumor_response_mean_df.pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
tumor_response_error_df = tumor_response_error_df.pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
display(tumor_response_mean_df)
display(tumor_response_error_df)
```

```python
#Creating the dataframes for the 4 treatments to be analyzed
tumor_study_mean_df = tumor_response_mean_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
tumor_study_error_df = tumor_response_error_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
```

```python
# Generate the Plot (with Error Bars)
tumor_study_mean_df.plot(kind='line', use_index=True, grid=True, yerr=tumor_study_error_df, marker='d', linewidth=0.5)
# Save the Figure

```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_mean_df = pd.DataFrame(study_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean())
meta_mean_df = meta_mean_df.reset_index()

# Preview DataFrame
meta_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_error_df = pd.DataFrame(study_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem())
meta_error_df = meta_error_df.reset_index()

# Preview DataFrame
meta_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_mean_df = meta_mean_df.pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')
meta_error_df = meta_error_df.pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')

# Preview that Reformatting worked
display(meta_mean_df)
display(meta_error_df)
```

```python
#Creating the dataframes for the 4 treatments to be analyzed
meta_study_mean_df = meta_mean_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
meta_study_error_df = meta_error_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
```

```python
# Generate the Plot (with Error Bars)
meta_study_mean_df.plot(kind='line', use_index=True, grid=True, yerr=meta_study_error_df, marker='d', linewidth=0.5)
# Save the Figure

# Show the Figure
plt.show()
```

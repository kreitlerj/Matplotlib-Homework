# Pymaceutical Analysis

This exercise was an analysis of data provided for a cancer study.  Out of ten treatments, we wanted to look at four specific ones.  The treatments being analyzed are Capomulin, Infubinol, Ketapril, and Placebo.  Between these treatments, only one showed a decrease of tumor volume through the 45 day trial period and that was Capomulin.  The other three rose at roughly the same rate.  All of the treatments had metastatic spread during the time period.  Capomulin had the slowest spread with only averaging 1.5 metastatic sites by the end of the trial.  Finally, Ketapril performed worse than the Placebo treatment.  This could indicate that the treatment could actually make the cancer worse.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
study_data = pd.merge(clinical_trial_data, mouse_drug_data, on='Mouse ID')

# Display the data table for preview
study_data.head()
```

## Check the Data for Errors

Make sure none of the mice are listed on more than one drug because that would indicate a recording error.

```python
#Check the to see if any mice are listed more than 10 times
#Each drug has 10 timepoints, so a mouse can be counted as alive more than that.
study_data['Mouse ID'].value_counts().head()
```

```python
#Check g989 to see if its listed for more than one drug
study_data_mice = study_data.set_index('Mouse ID')
mouse = study_data_mice.loc['g989', :]
mouse.head()
```

```python
#Delete the data for g989
clean_data_mice = study_data_mice.drop(index = 'g989')
clean_study_data = clean_data_mice.reset_index()
clean_study_data.head()
```

## Tumor Response to Treatment

```python
# Create dataframe to store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_response_mean_df = pd.DataFrame(clean_study_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean())
tumor_response_mean_df = tumor_response_mean_df.reset_index()
# Preview DataFrame
tumor_response_mean_df.head()
```

```python
# Create a dataframe and store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_response_error_df = pd.DataFrame(clean_study_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem())
tumor_response_error_df = tumor_response_error_df.reset_index()
# Preview DataFrame
tumor_response_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_response_mean_df = tumor_response_mean_df.pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
tumor_response_error_df = tumor_response_error_df.pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
# Preview that Reformatting worked
display(tumor_response_mean_df)
display(tumor_response_error_df)

tumor_response_mean_df.dtypes
```

```python
#Creating the dataframes for the 4 treatments to be analyzed
tumor_study_mean_df = tumor_response_mean_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
tumor_study_error_df = tumor_response_error_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
```

```python
# Generate the Plot (with Error Bars)
tumor_study_mean_df.plot(kind='line', use_index=True, grid=True, yerr=tumor_study_error_df, marker='d', linewidth=0.5)

plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

#Save Figure
plt.savefig('plots/Tumor Response.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_mean_df = pd.DataFrame(clean_study_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean())
meta_mean_df = meta_mean_df.reset_index()

# Preview DataFrame
meta_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_error_df = pd.DataFrame(clean_study_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem())
meta_error_df = meta_error_df.reset_index()

# Preview DataFrame
meta_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_mean_df = meta_mean_df.pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')
meta_error_df = meta_error_df.pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')

# Preview that Reformatting worked
display(meta_mean_df)
display(meta_error_df)
```

```python
#Creating the dataframes for the 4 treatments to be analyzed
meta_study_mean_df = meta_mean_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
meta_study_error_df = meta_error_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
```

```python
# Generate the Plot (with Error Bars)
meta_study_mean_df.plot(kind='line', use_index=True, grid=True, yerr=meta_study_error_df, marker='d', linewidth=0.5)

plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Metastatic Sites')

# Save the Figure
plt.savefig('plots/Metastatic Spread.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_df = pd.DataFrame(clean_study_data.groupby(['Drug', 'Timepoint'])['Mouse ID'].count())
survival_df = survival_df.reset_index()
survival_df = survival_df.rename(columns={'Mouse ID': 'Mouse Count'})

# Preview DataFrame
survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_summary_df = survival_df.pivot(values='Mouse Count', index='Timepoint', columns='Drug')

# Preview the Data Frame
survival_summary_df
```

```python
#Creating the dataframes for the 4 treatments to be analyzed
survival_study_df = survival_summary_df[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]

#Changing the data in the data frame from count to percentage - grab the initial mice count at the start
total_mice = survival_study_df.loc[0,:]

#For loop to calculate the percentages at different time points for each treatment to be evaluated
for column in survival_study_df:
    survival_study_df[column] = survival_study_df[column] / total_mice.loc[column] * 100 #dividing by the initial mice count
    
#Preview the dataframe with percentages
survival_study_df
```

```python
# Generate the Plot (Accounting for percentages)
survival_study_df.plot(kind='line', use_index=True, grid=True, marker='d', linewidth=1)

plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rates (%)')

# Save the Figure
plt.savefig('plots/Survival Rates.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = 100 * (tumor_response_mean_df.loc[45, :] - tumor_response_mean_df.loc[0, :]) / tumor_response_mean_df.loc[0, :] 
percent_change
```

```python
#Create a data frame with the treatments for the study and add a column to determine if the treatment was passing
study_percent_change = pd.DataFrame({'Percent Change': percent_change[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]})
study_percent_change['Passing'] = study_percent_change['Percent Change'] < 0 #passing condition: tumor shrank
study_percent_change

#Create a bar graph
x = study_percent_change.index.values
y = study_percent_change['Percent Change'].values

fig, ax = plt.subplots()
summary = ax.bar(x, y, color=study_percent_change['Passing'].map({True: 'g', False: 'r'}))

#format
ax.grid()
ax.set_ylim(-30, 100)
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')

# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(summary)

# Save the Figure
plt.savefig('plots/Percent Tumor Change.png')

# Show the Figure
plt.show(summary)
```
